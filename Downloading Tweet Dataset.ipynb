```python
import os
import tweepy as tw
import pandas as pd
import time
```

```python
#import data
data_filename = "./us-house/us-house/data/us-house.csv"
representatives = pd.read_csv(data_filename, error_bad_lines=False)
```

```python
representatives.columns
```

```python
consumer_key= ''
consumer_secret= ''
access_token= ''
access_token_secret= ''
```

```python
auth = tw.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tw.API(auth, wait_on_rate_limit=True)
```

```python
tweet_collection = pd.DataFrame()
```

```python
tweet_collection
```

```python
userID = "joebiden"
```

```python
rawTweets = api.user_timeline(screen_name=userID, 
                       # 200 is the maximum allowed count
                       count=200,
                       include_rts = False,
                       # Necessary to keep full_text 
                       # otherwise only the first 140 words are extracted
                       tweet_mode = 'extended'
                       )
tweets = [[userID, tweet.created_at, tweet.full_text] for tweet in rawTweets]
user_tweets = pd.DataFrame(data=tweets)
user_tweets.columns = ['user', 'datetime', 'content']
tweet_collection = tweet_collection.append(user_tweets)
```

```python
tweet_collection
```

```python
for index, row in representatives.iterrows():
    userID = row['twitter_handle']
    time.sleep(1)
    try:
        rawTweets = api.user_timeline(screen_name=userID, 
                               # 200 is the maximum allowed count
                               count=200,
                               include_rts = False,
                               # Necessary to keep full_text 
                               # otherwise only the first 140 words are extracted
                               tweet_mode = 'extended'
                               )
        tweets = [[userID, tweet.created_at, tweet.full_text] for tweet in rawTweets]
        user_tweets = pd.DataFrame(data=tweets)
        tweet_collection = tweet_collection.append(user_tweets)
        print("finished with " + userID)
        
    except tw.TweepError as e:
        if e == "[{u'message': u'Rate limit exceeded', u'code': 88}]":
            time.sleep(60*5) #Sleep for 5 minutes
        else:
            print (e)
            pass
```

```python
tweet_collection.to_csv(r'./tweets.csv',index=False)
```
